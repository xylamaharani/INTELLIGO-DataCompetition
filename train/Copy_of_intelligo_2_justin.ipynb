```python
df_order['order_purchase_timestamp']=pd.to_datetime(df_order['order_purchase_timestamp'])

# Create new columns for year and month
df_order['order_year'] = df_order['order_purchase_timestamp'].dt.year
df_order['order_month'] = df_order['order_purchase_timestamp'].dt.month
df_order
```

```python
df_payments=pd.read_csv('df_Payments.csv')
df_payments
```

```python
df_products=pd.read_csv('df_Products.csv')
df_products
```

```python
df=pd.merge(df_order,df_orderitems, on=['order_id'],how='inner')
df=pd.merge(df,df_customer, on=['customer_id'],how='inner')
df=pd.merge(df,df_payments, on=['order_id'],how='inner')
df_products = df_products.drop_duplicates(subset=['product_id'])
df_products=df_products.reset_index(drop=True)
df_products.info()
df=pd.merge(df,df_products, on=['product_id'],how='inner')
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
df['order_delivered_timestamp'] = pd.to_datetime(df['order_delivered_timestamp'])
df['order_delivered_timestamp'] = df['order_delivered_timestamp'].dt.strftime('%Y-%m-%d')
df_tes=df.dropna()
df_tes['order_delivered_timestamp'] = pd.to_datetime(df_tes['order_delivered_timestamp'])
df_tes['is_late'] = df_tes.apply(lambda row: 1 if row['order_delivered_timestamp'] > row['order_estimated_delivery_date'] else 0, axis=1)
df_tes['late'] = df_tes.apply(lambda row: (row['order_estimated_delivery_date'] - row['order_delivered_timestamp']).days, axis=1)
```

```python
import warnings
import optuna
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from sklearn.preprocessing import OneHotEncoder
import xgboost as xgb
from sklearn.compose import ColumnTransformer

warnings.simplefilter(action='ignore', category=FutureWarning)

X = df_tes[['product_weight_g','product_length_cm','product_height_cm','product_width_cm','customer_city', 'customer_state','order_year','order_month']]
y = df_tes['late']

categorical_features = ['customer_city', 'customer_state']

# Create a ColumnTransformer to apply one-hot encoding to categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'
)

# Fit and transform the training data
X = preprocessor.fit_transform(X)

# Create DMatrix objects for training and testing data
dmatrix = xgb.DMatrix(X, label=y)

def objective(trial):
    params = {
        'tree_method': 'hist',
        'device': 'cuda',
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.5),
        'subsample': trial.suggest_float('subsample', 0.5, 1),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.5, 1),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 30),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'alpha': trial.suggest_float('reg_alpha', 0, 1),
        'lambda': trial.suggest_float('reg_lambda', 0, 1),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 10),
        'max_delta_step': trial.suggest_float('max_delta_step', 0, 10),
    }

    best_xgb = XGBRegressor(**params)

    cv = KFold(n_splits=5, shuffle=True, random_state=42)

    # Perform cross-validation
    mse_scorer = make_scorer(score_func=mean_squared_error, greater_is_better=False)
    mse_scores = cross_val_score(best_xgb, X, y, cv=cv, scoring=mse_scorer)

    # Compute and return the mean MSE
    mean_mse = np.mean(mse_scores)
    return mean_mse

# Example of how to use Optuna to optimize the objective function
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)

# Get the best parameters
best_params = study.best_params
print(f'Best Parameters: {best_params}')

# Get the best score
best_score = study.best_value
print(f'Best Mean Squared Error: {best_score}')
```
